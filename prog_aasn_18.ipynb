Question 1
Create a function that takes a list of non-negative integers and strings and return a new list
without the strings.

```python
def filter_list(list):
    l=[]
    for char in list:
        if type(char) != str:
            l.append(char)
        else:
            continue
    return l
```

Question 2
The &quot;Reverser&quot; takes a string as input and returns that string in reverse order, with the
opposite case.

```python
def reverse(string):
    l=[]
    for char in string:
        if char.islower()==True:
            char=char.upper()
            l.append(char)
        else:
            char=char.lower()
            l.append(char)
    l=l[::-1]
    k="".join(l)
    print(k)
    
        
```

Question 4
Write a function that calculates the factorial of a number recursively.

```python
def factorial(n):
    var=1
    l=[]
    for char in range(n):
        l.append(char)
    for str in range(len(l),1,-1):
        var=str*var
    return var
        
        
        
```
