Question3
Create a function that takes a string and returns a string with its letters in alphabetical order.

```python
def alphabet_soup(string):
    l=[]
    l1=[]
    for char in string:
        l.append(ord(char))
        l.sort()
    for var in l:
        l1.append(chr(var))
    k="".join(l1)
    return k
        
```
